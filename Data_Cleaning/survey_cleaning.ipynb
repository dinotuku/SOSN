## Unify the schema of each year to make it easier to compare

```python
# import the cleaning util functions
from survey_cleaning_helper import *
```

```python
# Map hobby into a yes / no field, fill na with no
survey2016.loc[:, 'Hobby'] = map_hobby(survey2016['Hobby'], 2016)
survey2017.loc[:, 'Hobby'] = map_hobby(survey2017['Hobby'], 2017)
survey2018.loc[:, 'Hobby'] = map_hobby(survey2018['Hobby'], 2018)
survey2019.loc[:, 'Hobby'] = map_hobby(survey2019['Hobby'], 2019)
```

```python
# change the country field into iso3 format for easier comparison with the stackoverflow data
# fill Others / na with 'NA'
survey2016.loc[:, 'Country'] = map_country(survey2016['Country'])
survey2018.loc[:, 'Country'] = map_country(survey2018['Country'])
survey2019.loc[:, 'Country'] = map_country(survey2019['Country'])

# 2017 requires separate cleaning as some of the country name may not present in the coco utils
survey2017.loc[:, 'temp_coun'] = map_country(survey2017['Country']) 
survey2017.loc[survey2017.Country == 'Moldavia', 'temp_coun'] = 'MDA'
survey2017.loc[survey2017.Country == 'Azerbaidjan', 'temp_coun'] = 'AZE'
survey2017.loc[survey2017.Country == 'Tadjikistan', 'temp_coun'] = 'TJK'
survey2017.loc[survey2017.Country == 'Netherlands Antilles', 'temp_coun'] = 'NLD'
survey2017.loc[survey2017.Country == 'Polynesia (French)', 'temp_coun'] = 'PYF'
survey2017['Country'] = survey2017['temp_coun']
survey2017.drop(columns='temp_coun', inplace=True)
```

```python
# the years of coding is mapped to 4 distinct category, i.e. '0 - 2 years', '2 - 5 years', 
# '6 - 10 years’ and '11+ years', as it is the most common abstraction for the 4 years combined
# fill na with 'NA'
survey2016.loc[:, 'YearsCoding'] = map_yearsCoding(survey2016['YearsCoding'], 2016)
survey2017.loc[:, 'YearsCoding'] = map_yearsCoding(survey2017['YearsCoding'], 2017)
survey2018.loc[:, 'YearsCoding'] = map_yearsCoding(survey2018['YearsCoding'], 2018)
survey2019.loc[:, 'YearsCoding'] = map_yearsCoding(survey2019['YearsCoding'], 2019)
```

```python
# map the job satisfaction into a 7-point scale from 1 to 7 inclusive
# fill na with 0
survey2016.loc[:, 'JobSatisfaction'] = map_jobSat(pd.DataFrame(survey2016.JobSatisfaction), 2016)
survey2017.loc[:, 'JobSatisfaction'] = map_jobSat(pd.DataFrame(survey2017.JobSatisfaction), 2017)
survey2018.loc[:, 'JobSatisfaction'] = map_jobSat(pd.DataFrame(survey2018.JobSatisfaction), 2018)
survey2019.loc[:, 'JobSatisfaction'] = map_jobSat(pd.DataFrame(survey2019.JobSatisfaction), 2019)
```

```python
# fill na in converted salary with 0
survey2016.loc[:, 'ConvertedSalary'] = clean_salary(survey2016['ConvertedSalary'])
survey2017.loc[:, 'ConvertedSalary'] = clean_salary(survey2017['ConvertedSalary'])
survey2018.loc[:, 'ConvertedSalary'] = clean_salary(survey2018['ConvertedSalary'])
survey2019.loc[:, 'ConvertedSalary'] = clean_salary(survey2019['ConvertedSalary'])
```

```python
# split the languages worked with and desire to work with field into lists
# empty list if na or does not enter anything
survey2016.loc[:, 'LanguageWorkedWith'] = split_languages(survey2016['LanguageWorkedWith'], 2016)
survey2017.loc[:, 'LanguageWorkedWith'] = split_languages(survey2017['LanguageWorkedWith'], 2017)
survey2018.loc[:, 'LanguageWorkedWith'] = split_languages(survey2018['LanguageWorkedWith'], 2018)
survey2019.loc[:, 'LanguageWorkedWith'] = split_languages(survey2019['LanguageWorkedWith'], 2019)

survey2016.loc[:, 'LanguageDesireNextYear'] = split_languages(survey2016['LanguageDesireNextYear'], 2016)
survey2017.loc[:, 'LanguageDesireNextYear'] = split_languages(survey2017['LanguageDesireNextYear'], 2017)
survey2018.loc[:, 'LanguageDesireNextYear'] = split_languages(survey2018['LanguageDesireNextYear'], 2018)
survey2019.loc[:, 'LanguageDesireNextYear'] = split_languages(survey2019['LanguageDesireNextYear'], 2019)
```

```python
# map gender into 4 distinct categories, i.e. Male, Female, Transgender and Prefer not to disclose
# fill na with 'Prefer not to disclose'
survey2016.loc[:, 'Gender'] = map_gender(survey2016['Gender'], 2016)
survey2017.loc[:, 'Gender'] = map_gender(survey2017['Gender'], 2017)
survey2018.loc[:, 'Gender'] = map_gender(survey2018['Gender'], 2018)
survey2019.loc[:, 'Gender'] = map_gender(survey2019['Gender'], 2019)
```

```python
# map the stackoverflowVisit and stackoverflowParticipate count to 'I have never visited Stack Overflow (before today)', 
# 'Less than once per month or monthly', 'A few times per month or weekly', 'A few times per week', 
# 'Daily or almost daily', 'Multiple times per day' and 'NA'
survey2016.loc[:, 'StackOverflowVisit'] = clean_visit(pd.DataFrame(survey2016.StackOverflowVisit), 2016)
survey2017['StackOverflowVisit'] = clean_visit(pd.DataFrame(survey2017.loc[:, 'StackOverflowFoundAnswer':\
                                                                           'StackOverflowCopiedCode']), 2017)
survey2017.drop(columns=['StackOverflowFoundAnswer', 'StackOverflowCopiedCode'], inplace=True)
survey2018.loc[:, 'StackOverflowVisit'] = clean_visit(pd.DataFrame(survey2018.StackOverflowVisit), 2018)
survey2019.loc[:, 'StackOverflowVisit'] = clean_visit(pd.DataFrame(survey2019.StackOverflowVisit), 2019)

survey2016['StackOverflowParticipate'] = 'NA'
survey2017['StackOverflowParticipate'] = clean_participate(pd.DataFrame(survey2017.loc[:, 'StackOverflowNewQuestion':\
                                                                           'StackOverflowAnswer']), 2017)
survey2017.drop(columns=['StackOverflowNewQuestion', 'StackOverflowAnswer'], inplace=True)
survey2018.loc[:, 'StackOverflowParticipate'] = clean_participate(pd.DataFrame(survey2018.StackOverflowParticipate), 2018)
survey2019.loc[:, 'StackOverflowParticipate'] = clean_participate(pd.DataFrame(survey2019.StackOverflowParticipate), 2019)
```

```python
# add the year entry for the dataframe
survey2016['Year'] = 2016
survey2017['Year'] = 2017
survey2018['Year'] = 2018
survey2019['Year'] = 2019
```

```python
# keep only the age entry of the year 2019
survey2016['Age'] = 0
survey2017['Age'] = 0
survey2018['Age'] = 0
```
